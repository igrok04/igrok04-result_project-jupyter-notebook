```python
import numpy as np
import scipy
from scipy.optimize import fsolve
from matplotlib import pyplot as plt
```

## Дана функция:

```python

def f(x):
    return -12*(x ** 4) * np.sin(np.cos(x)) -18 * (x **3) + 5*(x**2) + 10*x - 30
```

## 1. Определить корни

```python
print("Корни:\n",scipy.optimize.fsolve(f, [-100, 100]))
```

## 2. Найти интервалы, на которых функция возрастает

```python
x = np.arange(-100, 100, 0.001)
start = None
end = None

print("Функция возрастает на отрезках: ")
for a in range(1, len(x)):
    # print(f(x[a]) > f(x[a-1]))
    if f(x[a]) > f(x[a-1]):
        if start is None:
            start = x[a-1]
    else:
        if start is not None:
            end = x[a-1]
            print("[", start, ", ", end, "] ")
            start = None

if start and not end:
    print("[", start, ", ", x[len(x)-1],"]")
```
